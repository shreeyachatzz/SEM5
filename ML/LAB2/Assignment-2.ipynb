Common categorical data types are:

*   boolean - Can take only two values like in true or false, male or female
*   nominal - Can take more than two values like in American - European - Asian - African
*   ordinal - Can take more than two values but the values have a logical order like in few - some - many

```python
# example of replacing nan values with mean values

data_notnull=data_reduced[data_reduced['YearlyIncome'].notnull()]
data_notnull.head()

mean_income=data_notnull['YearlyIncome'].mean()
print(mean_income)
data_reduced['YearlyIncome']=data_reduced['YearlyIncome'].replace(to_replace = np.nan, value = mean_income)
data_reduced.head()
```

```python
# To include only the object columns in our dataframe
obj_data_reduced = data_reduced.select_dtypes(include=['object']).copy()
obj_data_reduced.head()
```

```python
# To include other than object columns in our dataframe
nonobj_data_reduced = data_reduced.select_dtypes(exclude=['object']).copy()
nonobj_data_reduced.head()
```

```python
nonobj_data_reduced.info()
```

**Normalization is the process of scaling individual samples to have unit norm.**

The function normalize provides a quick and easy way to perform this operation on a single array-like dataset, either using the l1, l2, or max norms


**norm{â€˜l1â€™, â€˜l2â€™, â€˜maxâ€™}, default=â€™l2â€™**

The norm to use to normalize each non zero sample (or each non-zero feature if axis is 0).

**axis{0, 1}, default=1**

Define axis used to normalize the data along. If 1, independently normalize each sample, otherwise (if 0) normalize each feature.

```python
from sklearn.preprocessing import normalize
test_pd=pd.read_csv('test.csv')
test_pd_normalize=normalize(test_pd, norm='max', axis=0) #max normalization method
test_pd_normalize=pd.DataFrame(test_pd_normalize)
test_pd_normalize.head(10)
```

```python
from sklearn.preprocessing import normalize
nonobj_data_normalize=normalize(nonobj_data_reduced, norm='max', axis=0)
nonobj_data_normalize=pd.DataFrame(nonobj_data_normalize)
nonobj_data_normalize.head()
```
